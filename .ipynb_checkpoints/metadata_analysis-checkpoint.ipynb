```python
import os
import json
import pandas as pd

def load_json_files_from_directory(directory):
    json_files_data = {}
    
    # Walk through the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.json'):
                file_path = os.path.join(root, file)
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        json_data = json.load(f)
                        json_files_data[file_path] = pd.json_normalize(json_data)
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON from {file_path}: {e}")
                except IOError as e:
                    print(f"Error reading {file_path}: {e}")
    
    return json_files_data

    
directory_path = './Hackthon_JDOPCenter_Data/Farm1_Jennewein'
all_json_data = load_json_files_from_directory(directory_path)

def build_ohe_df (file_name, file_data):
    ohe_cols = {x:1 for x in list(file_data.columns)}
    return pd.DataFrame(ohe_cols, index=[file_name])
    
# Print the loaded JSON data
meta_columns = pd.concat([build_ohe_df(file_name, file_data) for file_name, file_data in all_json_data.items()]).fillna(0)
meta_columns
```
